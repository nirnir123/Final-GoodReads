```python
count_vec = CountVectorizer(stop_words=stop_wo,ngram_range=(1, 2), max_features= 20)
count_vec.fit(df1['description'])
text_vectorized = pd.DataFrame(count_vec.transform(df1['description']).toarray(),
                               columns=count_vec.get_feature_names())
text_vectorized
```

```python
df = df1.merge(text_vectorized,left_index=True, right_index=True)
df
```

```python
popular_words=df['dark']+df['even']+df['ever']+df['family']+df['find']+df['first']+df['life']+df['love']+df['magic']+df['man']+df['must']+df['never']+df['new']+df['old']+df['power']+df['time']+df['war']+df['world']+df['years']+df['young']
df3=df.copy()
df3['popular_words'] = popular_words
df3.columns
```
